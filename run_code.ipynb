```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

df = pd.read_csv('grid_nordeste.csv')


df_final = gpd.GeoDataFrame(
    df,
    geometry=gpd.GeoSeries.from_wkt(df['geometry']))

fig, ax = plt.subplots(1,1, figsize=(20, 10))
df_final.plot(color='red',markersize=15, ax=ax,label='Grid Nordeste')
ax.legend()
```

```python
import os
import glob
import json
import requests
import geopandas as gpd
import pandas as pd
import numpy as np
from shapely import wkt
import folium as fl

locations = list()
pb_grid['center_point'] = [ i.wkt for i in pb_grid['center_point']]
pb_grid['center_point'] = gpd.GeoSeries.from_wkt(pb_grid['center_point'])

for point in pb_grid['center_point']:
    locations.append(
        (
            point.y,
            point.x
        )
    )
```

```python
latitudes = [j for i, j in locations]
longitudes = [i for i, j in locations]

map = fl.Map(location=[-7.129694, -35.867750], zoom_start=7, tiles='OpenStreetMap')

for lat, long in zip(latitudes,longitudes):
    fl.Marker(
        location=[
            long,
            lat
        ],
        icon=fl.Icon(color="red"),
        popup='Ponto Nasa',
        name='Ponto Nasa',
        tooltip=f'<strong>Latitude:</strong> {lat} <br><strong>Longitude:</strong> {long}'
    ).add_to(map)

map
```
